### Step 1: Read the `neighborhood_coordinates.csv` file from the `Resources` folder into the notebook, and create a DataFrame named `neighborhood_locations_df`. Be sure to set the `index_col` of the DataFrame as “Neighborhood”.

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(Path('Resources/neighborhoods_coordinates.csv'))

# Review the DataFrame
neighborhood_locations_df
```

```python
neighborhood_locations_df.loc[neighborhood_locations_df['Neighborhood']=='Telegraph Hill']
```

### Step 2: Using the original `sfo_data_df` Dataframe, create a DataFrame named `all_neighborhood_info_df` that groups the data by neighborhood. Aggregate the results by the `mean` of the group.

```python
# Calculate the mean values for each neighborhood
all_neighborhood_info_df = sfo_data_df.groupby('neighborhood').mean().reset_index().drop('year',axis=1)

# Review the resulting DataFrame
all_neighborhood_info_df
```

### Step 3: Review the two code cells that concatenate the `neighborhood_locations_df` DataFrame with the `all_neighborhood_info_df` DataFrame. 

Note that the first cell uses the [Pandas concat function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html) to create a DataFrame named `all_neighborhoods_df`. 

The second cell cleans the data and sets the “Neighborhood” column. 

Be sure to run these cells to create the `all_neighborhoods_df` DataFrame, which you’ll need to create the geospatial visualization.

```python
# Using the Pandas `concat` function, join the 
# neighborhood_locations_df and the all_neighborhood_info_df DataFrame
# The axis of the concatenation is "columns".
# The concat function will automatially combine columns with
# identical information, while keeping the additional columns.
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
all_neighborhoods_df.loc[all_neighborhoods_df['Neighborhood']=='Telegraph Hill']
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
all_neighborhoods_df = all_neighborhoods_df.set_index('Neighborhood').dropna()

# Rename the "index" column as "Neighborhood" for use in the Visualization

all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"}).reset_index()

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

### Step 4: Using hvPlot with GeoViews enabled, create a `points` plot for the `all_neighborhoods_df` DataFrame. Be sure to do the following:

* Set the `geo` parameter to True.
* Set the `size` parameter to “sale_price_sqr_foot”.
* Set the `color` parameter to “gross_rent”.
* Set the `frame_width` parameter to 700.
* Set the `frame_height` parameter to 500.
* Include a descriptive title.

```python
from bokeh.models import HoverTool
hover = HoverTool(tooltips=[('Neighborhood','@Neighborhood'),
                            ('sale_price_sqr_foot','@sale_price_sqr_foot'),
                            ('gross_rent','@gross_rent{0.0}')])
hover
```

```python
# Create a plot to analyze neighborhood info
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size='sale_price_sqr_foot',
    color='gross_rent',
    frame_width=700,
    frame_height=500,
    title='San Fransisco Neighborhood Average Housing Prices 2010-2016',
    tiles='OSM',
    legend='top',
    hover_cols=['Neighborhood','sale_price_sqr_foot','gross_rent'],
    tools=[hover]
)
```

## Compose Your Data Story

Based on the visualizations that you have created, compose a data story that synthesizes your analysis by answering the following questions:

```python
# create a dataframe with calculated rent_sale_ratio 
prices_by_year_by_neighborhood_df = prices_by_year_by_neighborhood.reset_index()
prices_by_year_by_neighborhood_df['rent_sale_ratio'] = ( prices_by_year_by_neighborhood_df['gross_rent']
                                                         / ( prices_by_year_by_neighborhood_df['sale_price_sqr_foot']*12) )
prices_by_year_by_neighborhood_df = prices_by_year_by_neighborhood_df.drop(['sale_price_sqr_foot','gross_rent'],axis=1)
prices_by_year_by_neighborhood_df['neighborhood'].unique().size
```

```python
# set the index to year, neighborhood
ratios_df = prices_by_year_by_neighborhood_df.set_index(['year','neighborhood'])
ratios_df[ratios_df['rent_sale_ratio']>8.9]
```

```python
# transform the dataframe to have the columns as the rent-sale-ratio of neighborhoods, index by year

year_got = 0
df = pd.DataFrame()
df0 = df
ratios_df1 = ratios_df.reset_index()
for item in ratios_df.iterrows():
    if item[0][0] != year_got and year_got != 0:
        #print(f'break {item[0][0]} {year}')
        #display(df0)        
        df = pd.concat([df,df0],axis=0)
        df0 = pd.DataFrame()
    year_got = item[0][0]
    #print(item[0][0],item[0][1])
    df1 = ratios_df1[(ratios_df1['neighborhood']==item[0][1]) & (ratios_df1['year']==item[0][0])].set_index('year')
    df1 = df1.drop('neighborhood',axis=1).rename(columns={'rent_sale_ratio':item[0][1]})
    #display(df1)
    df0 = pd.concat([df0,df1],axis=1,join='outer')
    
df = pd.concat([df,df0],axis=0)

# plot the transformed dataframe
df.hvplot(
    ylabel='Rent-to-sale ratio',
    frame_width = 700,
    frame_height = 500,
    title='San Fransisco Neighborhoods Housing Rent-to-sale Ratio by Year',
    legend=False
)
```

**The Story**
On the geo-map plot we should look for small deep color circles, which indicates high rental gross low sale price, relatively.

Our purpose is to identify profitable rental houses to invest. Although the usual way of measure is price-to-rent ratio, \
(Sale Price)/((Monthly Rental)*12), in our dataset the sale price is per square feet, while the rental gross is not. No worries. \
With the plotted rent-to-price ratio we can clearly see which neighborhood is the most worthy investment. The trend can be also \
a considerable factor for future prediction.
